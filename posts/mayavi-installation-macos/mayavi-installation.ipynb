# Install mayavi on macOS

## System info

macOS Ventura, Version 13.1

Apple M1 chip

Anaconda version 2021.05 with python3.8.8

## Installation

```shell
pip install mayavi
pip install PyQt5
```

An error message emerges that demands for PyQt5<5.13 in order to be compatible with Spyder. Therefore, just run

```shell
pip uninstall PyQt5
pip install PyQt5==5.12
```

and it's done! To test the installation, run

```shell
mayavi2
```

This will open a UI of mayavi application. To use mayavi in jupyter lab/notebook, do the following:

```shell
jupyter nbextension install --py mayavi --user
jupyter nbextension enable --py mayavi --user

```

Below is a test demo.

````python
from mayavi import mlab
mlab.init_notebook('png')
s = mlab.test_plot3d()
s
````
